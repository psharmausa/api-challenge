```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
from scipy.stats import linregress
from datetime import datetime
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
# from git bash run this command-   'pip install citipy'
from citipy import citipy


# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
#create null variables
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=2000)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=2000)
lat_lngs = zip(lats, lngs)
#for testing purpses check how zip displays value-- delete when you sumbit
#print(tuple(lat_lngs))

# Identify nearest city for each lat, lng combination

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
 # If the city is unique, then add it to a our cities list
    if city not in cities:
         cities.append(city)
        
# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#testing-delete
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build query URL--delete
query_url = f"{url}appid={weather_api_key}&units={units}&q=Paris"
resp = requests.get(query_url).json()
resp
```

```python
#create url for query with imperial units and city at the end
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&"

# ASK SEAN / FELIPE HOW TO DERIEVE THE SET (#)- 
# Lists for variables
city_name = []
lat = []
lon = []
maxtemp = []
humidity = []
cloudiness = []
windSpeed = []
country = []
date = []

#print as requested
print("Beginning Data Retrieval     ")
print("-----------------------------")


#loop
i = 1
for city in cities:
    #print(cities)
    city=city
    query_url = url + "appid=" + weather_api_key + "&q=" + city
    response = requests.get(query_url).json()

#extract data. Missing data is handled by try and except  
    try:
        city_name.append(response["name"])
        lon.append(response["coord"]["lon"])
        lat.append(response["coord"]["lat"]) 
        maxtemp.append(response["main"]['temp_max'])
        cloudiness.append(response["clouds"]["all"])
        humidity.append(response["main"]['humidity'])
        windSpeed.append(response["wind"]["speed"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
    
        print(f"Processing record {i}|{city}")
        i+=1
    except (KeyError):
        print("City not found. Skip!!.")
        pass
              
# complete 
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_df=pd.DataFrame({"City": city_name,
                           "Country": country,
                           "Temperature": maxtemp,
                           "Humidity": humidity,
                           "Cloudiness": cloudiness,
                           "Windspeed":windSpeed,
                           "Latitude": lat,
                           "Longitude": lon,
                           "Date": date
                          })
weather_df
```

```python
# save the data to csv file
output_file = "../output_data/cities.csv"
weather_df.to_csv(output_file, index=False)
weather_df
```

```python
#Northern Hemisphere (greater than or equal to 0 degrees latitude)and
#Southern Hemisphere (less than 0 degrees latitude)

# linear regression variables for lat > 0 
north_df = weather_df.loc[weather_df["Latitude"] > 0]
north_lat=north_df["Latitude"]
north_temp = north_df['Temperature']
north_humid = north_df['Humidity']
north_cloud = north_df['Cloudiness']
north_windspeed = north_df['Windspeed']

# linear regression variables for lat < 0 
south_df = weather_df.loc[weather_df["Latitude"] < 0]
south_lat=south_df["Latitude"]
south_temp = south_df['Temperature']
south_humid = south_df['Humidity']
south_cloud = south_df['Cloudiness']
south_windspeed = south_df['Windspeed']
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python

# Add the linear regression equation and line to plot
# x_values = housing_data['RM']
# y_values = housing_data['MEDV']
# (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
# regress_values = x_values * slope + intercept
# line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# plt.scatter(x_values,y_values)
# plt.plot(x_values,regress_values,"r-")
# plt.annotate(line_eq,(6,10),fontsize=15,color="red")
# plt.xlabel('Rooms in House')
# plt.ylabel('Median House Prices ($1000)')

#set data to plot
#x,y value
north_lat=north_df["Latitude"]
north_temp = north_df['Temperature']

(slope, intercept, rvalue, pvalue, std_err) = linregress(north_lat, north_temp)
regress_value = slope * north_lat + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(line_eq)
print(f"The r-value is:  {rvalue}")


plt.scatter(north_lat, north_temp)
plt.plot(north_lat,regress_value,"r-")
plt.annotate(line_eq,(30,100),fontsize=15,color="red")

#add titles and axis labels
plt.xlabel("Latitude")
plt.ylabel("Temperature")
plt.title("Northern Hemisphere - Max Temp vs. Latitude Linear Regression")

#save it in file
plt.savefig("../output_data/N_Hemp_Temp.png")
plt.show()
```
